```python
# Task 1. Input Modeling (Graph + Data)
```

```python
import networkx as nx

# Graph (distance matrix as edges)
G = nx.Graph()

edges = [
    (0, 1, 10), (0, 2, 15), (0, 3, 20),
    (1, 2, 35), (1, 3, 25),
    (2, 3, 30)
]

G.add_weighted_edges_from(edges)

# Parcel data
parcels = [
    {"id": 1, "value": 100, "weight": 5, "time": 2},
    {"id": 2, "value": 60, "weight": 3, "time": 1},
    {"id": 3, "value": 120, "weight": 7, "time": 3},
]

# Vehicle constraint
max_capacity = 10
```

```python
# Task 2. Recurrence-Based Route Cost (Recursive TSP Style)
```

```python
import math

def tsp_recursive(graph, current, visited):
    if len(visited) == len(graph.nodes):
        return graph[current][0]['weight']  # return to start

    min_cost = math.inf
    for neighbor in graph.nodes:
        if neighbor not in visited:
            cost = graph[current][neighbor]['weight'] + tsp_recursive(graph, neighbor, visited | {neighbor})
            min_cost = min(min_cost, cost)

    return min_cost

print("Minimum route cost:", tsp_recursive(G, 0, {0}))
```

```python
# Task 3: Greedy + DP (Knapsack Style)
```

```python
# Greedy (Value/Weight)
```

```python
parcels_sorted = sorted(parcels, key=lambda x: x['value']/x['weight'], reverse=True)

selected = []
total_weight = 0

for p in parcels_sorted:
    if total_weight + p['weight'] <= max_capacity:
        selected.append(p)
        total_weight += p['weight']

print("Selected parcels (Greedy):", selected)
```

```python
# DP (Knapsack)
```

```python
def knapsack(parcels, capacity):
    n = len(parcels)
    dp = [[0]*(capacity+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for w in range(capacity+1):
            if parcels[i-1]['weight'] <= w:
                dp[i][w] = max(
                    parcels[i-1]['value'] + dp[i-1][w - parcels[i-1]['weight']],
                    dp[i-1][w]
                )
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

print("Max value (DP):", knapsack(parcels, max_capacity))
```

```python
mst = nx.minimum_spanning_tree(G)

print("MST edges:")
for edge in mst.edges(data=True):
    print(edge)
```

```python
# Task 5: TSP using Brute Force
```

```python
from itertools import permutations

def tsp_bruteforce(graph):
    nodes = list(graph.nodes)
    min_path = None
    min_cost = float('inf')

    for perm in permutations(nodes[1:]):
        path = [0] + list(perm) + [0]
        cost = 0

        for i in range(len(path)-1):
            cost += graph[path[i]][path[i+1]]['weight']

        if cost < min_cost:
            min_cost = cost
            min_path = path

    return min_path, min_cost

path, cost = tsp_bruteforce(G)
print("Optimal Path:", path)
print("Cost:", cost)
```
